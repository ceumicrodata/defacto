```python
import pandas
import os
import unicodecsv

RAW_DATA_DIR = '../data/download/'
V4_group = ['CZ','SK','PL','HU']
V3_group = ['CZ','SK','PL']
EU15_group = 'BE, DK, DE, IE, EL, ES, FR, IT, LU, NL, AT, PT, FI, SE, UK'.split(', ')

geo_to_plot = ['EU15','V3','HU']
geo_to_upload = ['EU15','V3','HU','CZ','SK','PL']

class Indicator(object):
    
    def __init__(self, short_name, long_name, csv_file=None, dataframe=None, number='1.1', topic=u'Egyéb'):
        self.topic = topic
        self.short_name = short_name
        self.long_name = long_name
        self.number = number.split('.')
        self.csv_file = csv_file
        self.dataframe = dataframe
        self.datetype = None
        if dataframe is None:
            self.read_eurostat_data(RAW_DATA_DIR)
        if self.csv_file is None:
            self.csv_file = self.short_name + '.csv'
        self.dateMin = self.dataframe.index.min().strftime('%Y-%m-%d')
        self.dateMax = self.dataframe.index.max().strftime('%Y-%m-%d')
        
    def preprocess_eurostat_dates(self, dataframe, min_year=1995):
        date_types = list(set(dataframe['date_type'].values))
        if len(date_types) > 1:
            if 'Q' in date_types:
                dataframe = dataframe[dataframe['date_type']=='Q']
            else:
                dataframe = dataframe[dataframe['date_type']=='M']
        dataframe = dataframe.drop('date_type',1)
        dataframe.set_index('date',inplace=True)
        min_year = str(max(dataframe.index.min().year, min_year))
        return dataframe[min_year:]
        
    def read_eurostat_data(self, raw_data_dir):
        # print self.csv_file
        rawdata = pandas.read_csv(raw_data_dir + self.csv_file, parse_dates=['date'])
        try:
            self.date_type = list(set(rawdata['date_type'].values))           
            self.dataframe = self.preprocess_eurostat_dates(rawdata)
        except:
            print self.csv_file + ' has not a eurostat type date'
            self.dataframe = rawdata
            
    def pretty_number(self):
        return '-'.join(['%s' % item for item in self.number])
                
                

def group_weighted_average(data, weight_group, weight_data):
    sum_of_weights = 0.0
    sumprod = 0.0
    for geo in weight_group:
        sum_of_weights += weight_data[geo]
        sumprod += weight_data[geo] * data[geo]
    return sumprod / sum_of_weights 
        
def plot_with_title(dataframe, title):
    return None
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.set_title(title)
    dataframe.plot(ax=ax)
    # Shink current axis's height by 10% on the bottom
    box = ax.get_position()
    ax.set_position([box.x0, box.y0,box.width * 0.8, box.height])
    # Put a legend below current axis
    ax.legend(loc='lower left', bbox_to_anchor=(1, 0.4))
    plt.show()
    
def generate_monthly_from_annual(df):
    min_year = df.index.min().year
    dr = pandas.date_range(datetime.date(min_year,1,1),df.index.max(),freq='M')
    df_m = pandas.DataFrame(index=dr, columns=df.columns)
    for index_value in df_m.index:
        list_of_valuses_for_year = [float(df[str(index_value.year)][geo]) for geo in df.columns]
        df_m.loc[index_value] = list_of_valuses_for_year
    return df_m

def interpolate_quaterly_from_annual(df):
    return df_q
```

# Load all data series

```python
def load_input_data(spec_file):
    input_dfs = []
    input_dfs_map = {}
    with open(spec_file, 'r') as file:
        csvreader = unicodecsv.DictReader(file, delimiter=';',
                                              encoding='iso-8859-1')
        index = 0
        for row in csvreader:
            csv_file = row['table_name'] + '.csv'
            #print row['indicator_name'],row['alapadat'],csv_file
            indicator = Indicator(row['indicator_name'],row['alapadat'],csv_file=csv_file)
            input_dfs.append(indicator)
            input_dfs_map.update({indicator.short_name:index})
            index += 1
    return input_dfs, input_dfs_map

input_indicators, input_indicator_map = load_input_data('./metadata/series_metadata.csv')

for i, indicator in enumerate(input_indicators):
    print u'{} {}:'.format(i+1,indicator.long_name)
    print '\tOriginal date type: {}'.format(indicator.date_type)
    #print indicator.dataframe.index.freq
    print '\tShort indicator name: {}'.format(indicator.short_name)
    print repr(indicator.dataframe.dropna(how='all').tail(15))
    print '\n\n' 

def get_input_data(short_indicator_name):
    return input_indicators[input_indicator_map[short_indicator_name]].dataframe
    
output_indicators = []
```

### 3.1 HICP infláció

```python
# Used eurostat input: 
hicp_base2005 = get_input_data('hicp_base2005')
# hicp_cow_eu25 = get_input_data('hicp_cow_eu25')
hicp_cow_eur = get_input_data('hicp_cow_emu')
hicp_cow_eu27 = get_input_data('hicp_cow_eu27')

for country in hicp_cow_eu27.columns:
    if country == 'EA' or country not in hicp_cow_eur.columns: continue
    NaN_vector = pandas.isnull(hicp_cow_eu27[country])
    hicp_cow_eu27[country][NaN_vector] = (hicp_cow_eu27[NaN_vector]['EA'] *
                                          hicp_cow_eur[NaN_vector][country] / 1000)
    
hicp_in_year = hicp_base2005 / (hicp_base2005[hicp_base2005.index.month==12].shift(1, freq='A').
                                reindex(index=hicp_base2005.index, method='bfill'))

hicp_cow_m = generate_monthly_from_annual(hicp_cow_eu27)

hicp_in_year['V3'] = group_weighted_average(hicp_in_year, V3_group, hicp_cow_m)
hicp_in_year['EU15'] = group_weighted_average(hicp_in_year, EU15_group, hicp_cow_m)
hicp_yearly = hicp_in_year[hicp_in_year.index.month==12]

hicp_base2005['V3'] = hicp_in_year['V3'] * (hicp_yearly['V3'].cumprod().shift(1, freq='A').
                                            reindex(index=hicp_base2005.index, method='bfill'))
hicp_base2005['EU15'] = hicp_in_year['EU15'] * (hicp_yearly['EU15'].cumprod().shift(1, freq='A').
                                                reindex(index=hicp_base2005.index, method='bfill'))

hicp_change = (hicp_base2005 / hicp_base2005.shift(12, freq='M') - 1) * 100

hicp_change.index.name = 'date'

plot_with_title(hicp_change[geo_to_plot],u'HIPC infláció')

output_indicators.append(Indicator('inflation',number='3.1', long_name=u'HIPC infláció', 
                                   dataframe=hicp_change[geo_to_upload].dropna(how='all'), topic=topic))
```

### 7.2 Hosszú lejáratú referencia államkötvényhozam

```python
# Used eurostat input:
gov_bond_yield_LT = get_input_data('gov_bond_yield_10Y')


#Súly:
nominal_GDP_q = get_input_data('gdp_eur_q_nsa')
nominal_GDP_arq_shift2 = (pandas.rolling_sum(nominal_GDP_q, 4, freq='Q')
            .shift(2, freq='Q'))
weights = nominal_GDP_arq_shift2.reindex(pandas.date_range(start=nominal_GDP_arq_shift2.index.min(),
                                                    end=nominal_GDP_arq_shift2.index.max(),
                                                    freq='M')).fillna(method='bfill')

gov_bond_yield_LT['V3'] = group_weighted_average(gov_bond_yield_LT, V3_group, weights)
gov_bond_yield_LT['EU15'] = group_weighted_average(gov_bond_yield_LT, EU15_group, weights)

plot_with_title(gov_bond_yield_LT[geo_to_plot + ['EA']],u'Hosszú lejáratú referencia államkötvényhozam')
output_indicators.append(Indicator('longtermyield', number='7.2', 
                             long_name=u'Hosszú lejáratú referencia államkötvényhozam',
                             dataframe=gov_bond_yield_LT[geo_to_upload].dropna(how='all'), topic=topic))
```

### 8.1 Munkaegységköltség-alapú effektív reálárfolyam (emelkedés = versenyképesség-javulás)

```python
# Used eurostat input:
real_ulc_exrate = get_input_data('real_ulc_exrate')
export_to_eu27_m = get_input_data('export_to_eu27')
print real_ulc_exrate.tail()
export_to_eu27_qrm = pandas.rolling_sum(export_to_eu27_m,3,freq='M')
export_to_eu27_qrm['CZ+PL+SK']=export_to_eu27_qrm['CZ']+export_to_eu27_qrm['PL']+export_to_eu27_qrm['SK']
export_to_eu27_qrm_share = export_to_eu27_qrm / export_to_eu27_qrm['CZ+PL+SK']
export_to_eu27_qrm_share = export_to_eu27_qrm_share.drop('CZ+PL+SK', axis=1)

ulc_competitiveness = real_ulc_exrate.apply(lambda x: 1/(x/100)*100)
ulc_competitiveness['V3'] = group_weighted_average(ulc_competitiveness, V3_group, export_to_eu27_qrm_share)

plot_with_title(ulc_competitiveness[['HU','V3']], u'Bérköltség-alapú versenyképesség\n(emelkedés = versenyképesség-javulás)')

output_indicators.append(Indicator('competitiveness', number='8.1',
                            long_name=u'Bérköltség-alapú versenyképesség (emelkedés = versenyképesség-javulás)',
                            dataframe=ulc_competitiveness.dropna(how='all'), topic=topic))
```
